### Issue #1:Underscore '_' present instead of space in dog breeds (p1,p2,p3) and a Few names with '-' present


#### Define:
1.Replace '_' with space in breed
2.Replace '-' with space/no space for breeds wherever needed based on actual name browsing
3.Convert the starting letter in breed from lower case to upper case

#### Code


```python
db = pd.read_csv('https://raw.githubusercontent.com/paiv/fci-breeds/main/fci-breeds.csv')
db.head()
```

Notice that the dog breed name in db dataframe are all in capital letters and seperated by space while the ones in our original dataframe(df2) is in small letters and seperated by '_', an underscore. Hence if we trying to do a matching check, it would not work until we re-format them to be uniform. We can either reformat db to look ike df2 or vice versa

```python
#Building a helper function to replace '_' with ' ' and convert all to uppercase

def refr (breed):
    
    breed = breed.replace('_', ' ')
    breed = breed.upper()
    
    return breed
```

```python
confd = []

for bn in range (len(df2_clean)):
    brd = df2_clean['p1'][bn]
    brd = refr(brd)
    if brd in db['name'].values.tolist():
      confd.append(brd)

    elif  brd in db['name'].values.tolist() == False and refr(df2_clean['p2'][bn]) in db['name'].values.tolist()==True:
      brd = refr(df2_clean['p2'][bn])
      confd.append(brd)

    elif brd in db['name'].values.tolist() == False and refr(df2_clean['p2'][bn]) in db['name'].values.tolist()==False and refr(df2_clean['p3'][bn]) in db['name'].values.tolist()==True:
      confd.append(refr(df2_clean['p3'][bn]))
    else:
      brd = 'nan'
      confd.append(brd)
```

```python
len(confd), df2.shape
```

```python
df2['p1'] = confd

df2.head()
```
